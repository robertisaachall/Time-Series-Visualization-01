```python
import pandas as pd
import matplotlib.pyplot as plt

#DATA RETRIEVED FROM THIS WEBSITE:    https://www.kaggle.com/htagholdings/property-sales , USED: raw_sales.csv file. 


df = pd.read_csv('data/raw_sales.csv')
df["datesold"] = df["datesold"].astype("datetime64")
df = df.set_index("datesold")
```

# Time series code for one bedroom houses

```python
one_bedroom_data = df.loc[(df["bedrooms"] == 1) & (df["propertyType"] == "house")]


plt.figure(figsize=(12,10))
plt.xlabel("Date")
plt.ylabel("Price of Home")
plt.title("Price of 1-bedroom Homes Time Series")
plt.ticklabel_format(style='plain')     
plt.bar(one_bedroom_data.index,one_bedroom_data["price"],width=30)
```
