```python
import pandas as pd
import matplotlib.pyplot as plt




df = pd.read_csv('data/COVID-19_Diagnostic_Laboratory_Testing__PCR_Testing__Time_Series.csv')
df["date"] = df["date"].astype("datetime64")
df = df.set_index("date")

df
```

# Code for Total New Reported for overall outcome

```python
plt.figure(figsize=(12,10))
plt.xlabel("Date")   
plt.ylabel("Results Reported")
plt.title("Overall Total New Reported Not Seperated into Outcome")
plt.plot(df.index,df["new_results_reported"])
```

# Code for negative data for overall outcome

```python
negative_data = df.loc[(df["overall_outcome"] == "Negative")]
plt.figure(figsize=(12,10))
plt.xlabel("Date")
plt.ylabel("Results Reported")
plt.title("Negative Outcome Total New Reported Overall")
plt.bar(negative_data.index,negative_data["new_results_reported"])
```

# Code for Positive data for overall outcome

```python
positive_data = df.loc[(df["overall_outcome"] == "Positive")]
plt.figure(figsize=(12,10))
plt.xlabel("Date")
plt.ylabel("Results Reported")
plt.title("Positive Outcome Total New Reported Overall")
plt.bar(positive_data.index,positive_data["new_results_reported"])
```

# Code for Inconclusive Data for overall new reported

```python
inconclusive_data = df.loc[(df["overall_outcome"] == "Inconclusive")]
plt.figure(figsize=(12,10))
plt.xlabel("Date")
plt.ylabel("Results Reported")
plt.title("Inconclusive Outcome Total New Reported Overall")
plt.bar(inconclusive_data.index,inconclusive_data["new_results_reported"])
```
